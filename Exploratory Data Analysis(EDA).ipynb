# Data Visualization

```python
plt.style.use('fivethirtyeight')
```

```python
#EV Makes
df['Make'].value_counts().plot(kind='bar', y = 'Count', figsize = (12,6))
```

```python
#Electric Vechile Type
df['Electric Vehicle Type'].value_counts().plot(kind='pie')
```

```python
df['Clean Alternative Fuel Vehicle (CAFV) Eligibility'].value_counts().plot(kind='bar')
```
